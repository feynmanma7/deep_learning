## 2.1 Sigmoid
> $\sigma(x) = \frac{1}{1+exp(-x)}$


```python
import matplotlib.pyplot as plt
import matplotlib.mlab as mlab
import math
```

```python
def sigmoid(x):
    return 1. / (1 + math.exp(-x))
```

```python
x = np.linspace(-20, 20, 1000)
y = list(map(lambda i : sigmoid(i), x))

plt.plot(x, y, 'b')
plt.axhline(y=1.0, ls='--', color = 'g')
plt.axhline(y=0.0, ls='--', color = 'g')
plt.axvline(x=0.0, ls='--', color = 'r')
plt.title('Sigmoid')

plt.show()
```


## 2.2 Tanh

> $tanh(x) = 2\sigma(x) - 1$


```python
def tanh(x): 
    return 2 * sigmoid(x) - 1
```

```python
x = np.linspace(-20, 20, 1000)
y = list(map(lambda i : tanh(i), x))

plt.plot(x, y, 'b')
plt.axhline(y=1.0, ls='--', color = 'g')
plt.axhline(y=-1.0, ls='--', color = 'g')
plt.axvline(x=0.0, ls='--', color = 'r')
plt.title('Tanh')

plt.show()
```


## 2.3 ReLU

> $ReLU(x) = max(0, x)$

```python
def relu(x):
    return max(0, x)
```

```python
x = np.linspace(-20, 20, 1000)
y = list(map(lambda i : relu(i), x))

plt.plot(x, y, 'b')
plt.axhline(y=0.0, ls='--', color = 'g')
plt.axvline(x=0.0, ls='--', color = 'r')
plt.title('ReLU')

plt.show()
```
